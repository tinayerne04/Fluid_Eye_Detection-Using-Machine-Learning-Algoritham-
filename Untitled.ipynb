```python
num_fluid_train = len(os.listdir(train_fluid))
num_normal_train = len(os.listdir(train_normal))
num_fluid_valid = len(os.listdir(validation_fluid))
num_normal_valid = len(os.listdir(validation_normal))
print('fluid training data ',num_fluid_train)
print('normal training Data ',num_normal_train)
print('fluid Valid data ',num_fluid_valid)
print('normal valid data ',num_normal_valid )
```

```python
total_train_data=num_fluid_train+num_normal_train
total_val_data=num_fluid_valid+num_normal_valid
print('Total Training images = ',total_train_data)
print('Total validation images = ',total_val_data)
```

```python
batch_size = 30
epochs = 100
IMG_HIEGHT = 150
IMG_WIDHT = 150
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255)
validation_image_generator = ImageDataGenerator(rescale=1./255)
```

```python
#flow_from_directory se data directory me stored hoga and batch me divide hoga
#for assining the label we use Class_mode
train_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,
                                                           directory=train_dir,
                                                           shuffle=True,target_size=(IMG_HIEGHT,IMG_WIDHT),
                                                           class_mode='binary')
```

```python
train_data_gen.class_indices
```

```python
#image genrator image ko data me convert krne keliye use hota hai
val_data_gen = validation_image_generator.flow_from_directory(batch_size=batch_size,
                                                              directory=validation_dir,
                                                              target_size=(IMG_HIEGHT,IMG_WIDHT),
                                                              class_mode='binary')
```

```python
def plotImages(images):
    fig, axes = plt.subplots(1, len(images), figsize=(28, 28))
    for img, ax in zip(images, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.show()
```

```python
plotImages(sample_image[10:15])
```

```python
model_new = Sequential([
    Conv2D(16,3,padding='same',activation='relu',input_shape=(IMG_HIEGHT,IMG_WIDHT,3)), #150
    MaxPooling2D(), #75
    Dropout(0.2),
    Conv2D(32,3,padding='same',activation='relu'), #75
    MaxPooling2D(),#37
    Dropout(0.2),
    Conv2D(64,3,padding='same',activation='relu'),#37
    MaxPooling2D(),#18
    Dropout(0.2),
    Flatten(),
    Dense(512,activation='relu'),
    Dropout(0.2),
    Dense(1,activation='sigmoid')
])
```

```python
model_new.summary()
```

```python
model_new.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model_new.fit(train_data_gen,steps_per_epoch=total_train_data//batch_size,epochs=100,validation_data=val_data_gen,validation_steps=total_val_data//batch_size)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(epochs)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(epochs_range,acc,label='Training Accuracy')
plt.plot(epochs_range,val_acc,label='Val Accuracy')
plt.legend(loc=0)
plt.title('Accuracy')
plt.subplot(1,2,2)
plt.plot(epochs_range,loss,label='Training Loss')
plt.plot(epochs_range,val_loss,label='Val LOss')
plt.legend(loc=0)
plt.title('Loss')
plt.show()
```

```python
class_name = ['Normal','Fluid']
#testing cascad
import cv2
import numpy as np
face_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
font = cv2.FONT_HERSHEY_SIMPLEX
# To capture video from webcam. 
cap = cv2.VideoCapture(0)
# To use a video file as input 
# cap = cv2.VideoCapture('filename.mp4')
while True:
    # Read the frame
    _, img = cap.read()

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    test = img
    test = cv2.resize(test,(150,150))
    test = test.reshape(1,150,150,3)
    #iden = []
    pred = model_new.predict(test)
    #iden.append(int(pred[0][0]))
    #print(iden)
    

    # Detect the faces
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    #print(faces.shape)
    # Draw the rectangle around each face
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img, str(class_name[int(y_pred[0][0])]), (x+5,y-5), font, 1, (0,255,0), 2)

    # Display
    cv2.imshow('img', img)

    # Stop if escape key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
# Release the VideoCapture object
cap.release()
cv2.destroyAllWindows()
```
